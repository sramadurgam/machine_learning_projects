# Mushroom Classification 
## Predict If Mushroom is Edible or Poisonous Using Decision Tree

### 1. Load Data and Encode

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
shroom = pd.read_csv('Data/mushrooms.csv')
shroom.sample(5)
```

```python
enc = LabelEncoder()
shroom_encoded = shroom.apply(enc.fit_transform)
shroom_encoded.sample(5)
```

### 2. Create Training and Testing Sets

```python
X = shroom_encoded.iloc[:,1:23].values
y = shroom_encoded.iloc[:,0:1].values
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.30)
X_test.shape
```

### 3. Train Model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'gini')
dt.fit(X_train, y_train)
```

### 4. Predict and Calculate Accuracy

```python
y_pred = dt.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test, y_pred))
```

### 5. Implementing Logistic Regression
#### Test if only Decision Tree algorithm gives 100% accuracy

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test, y_pred))
```
